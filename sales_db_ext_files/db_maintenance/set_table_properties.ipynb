```python
# set table properties

db_name = "sales_data.sales_db"

tables = spark.catalog.listTables(db_name)

count = 0

for tbl in tables:
    print(f"Applying updated properties to table: {tbl.name}")

    if "vw" in tbl.name:
        print(f"{tbl.name} is a view\nSkipping to next table")
        continue
    else:
        try:
            spark.sql(f"ALTER TABLE {db_name}.{tbl.name} SET TBLPROPERTIES ('delta.enableDeletionVectors' = true)")
            
            print("Table properties updated successfully")

            count += 1

        except Exception as e:
            print(f"Error applying properties to table {tbl.name}: {e}")

print(f"{count} tables updated successfully")
```
